## 6.1 Why model Calcium Concentration?
- In the most neural simulations, the concentration of the main charge carriers (Na, K, Cl) are not explicitly kept track of since they do not change much following typical neural membrane  current.
- Only the electrical charge they carry is modeled.
  - This is because the membrane voltage V is assumed to be the only important variable.

This inherently assumes that the slow-acting ion pumps working in the background keep the ion concentrations constant in time. If there is a lot of spikes, this is not nessecerally true (if there's too much potassium between the cells (a lot of AP's), and ion pumps cannot keep up, that can change the membrane potential and reversal potential etc.).
Sometimes you need to take ions into account. That is something we will talk about in the second block of this course.

**But calcium (Ca) is different**.
- Intracellular Ca concentration is very low ~ 100nM in the cell, rather than several mM like the others (everything else as well as $[Ca]_ {out}$).

This means that the calcium currents will affect the intracellular Ca concentration needed to reliably compute the Ca reversal potential:
$$
E_{Ca} = \frac{RT}{Z_{Ca}F}\ln\frac{[Ca]_ {out}}{[Ca]_ {in}}
$$

That means that changing $[Ca]_ {in}$, it changes the reversal potential of calcium. This is something that can happen on a short time scale.

#### Why is Calsium important?
Calsium also plays a role in many intracellular signaling pathways
- Acts as a **second messenger** in the cell, triggering processes such as:
  - Secretion of neurotransmitters at synapses.
  - Regulating gene expression (what determines how much an ion channel should make, not a very simple story). Have DNA which encodes proteins, but how is the amount of these you make regulated? There are homeostatic regulations.
    - "The degree to which a gene is being expressed"/"how much is made".
  - Synaptic plasticity

- **Calcium-Imaging**:
  - Eat something $\Rightarrow$ Measure light to get an idea of calcium in a cell
  - Slow temporal resolution. Calcium hangs in there for a long time.


## 6.2 Diffusion (and electrodiffusion), (6.1 and 6.3 in Sterratt)
- Diffusion is a key process for intracellular transport.
- Diffusion equation is as follows:
$$
\frac{\partial c}{\partial t} = D \nabla^2 c
$$

The solution of the diffusion equation with $N$ molecules (e.g Ca++ ions) starting at the origin $\vec{r}=0$ at a time $t=0$ is given by:
$$
c(\vec{r}, t)=\frac{N}{(4\pi Dt)^{3/2}}\cdot  e^{-r^2/(4\pi D)}
$$

or more simply:
$$
c(\vec{r},t)=\frac{N}{(\sqrt{2\pi}\sigma(t))^3}\cdot  e^{-r^2/(2\sigma^2)}
$$

with time, the radial distribution gets more and more spread out. Concentration in the middle goes down according to the formula. This is a Gaussian (normal) distribution with time dependent standard deviation $\sigma (t) = \sqrt{2Dt}$. $D$ is the diffusion constant, depends on the molecule and the surroundings.

To get intuition about what this means w.r.t calcium we can put in some numbers:
- $D_{Ca} \approx 2\cdot 10^{-10} m^2/s$

How far does it then typically diffuse in $1\mu s$, $10 ms$, or $1 s$?

- $1\mu s$: $ \ r_{drift}~\sqrt{2Dt}=\sqrt{2\cdot 2\cdot 10^{-10}\cdot 10^{-6}}m=0.02\mu m$
- $10ms \Rightarrow r_{drift} = 2\mu m$
- $1s \Rightarrow r_{drift} = 20\mu m$

Tells us that for $t=100s$, you get a mean drift of $200\mu m$.
#### This is a very slow process that doesnt take you very far. It is important however for short distance transport in the cells.
To speed up transport, the brain has come up with a **railway system** (tubule requires atp)
Recall that Ca++ is an ion, so is also transported by electric fields (Electrodiffusion).

# 6.3 Modeling Intracellular Calcium

$$
\frac{d}{dt}[Ca^{++}] = J_{drift} - J_{buff} + J_{CC} - J_{pump} - J_{up} + J_{rel} - J_{leak}
$$

- $J_{CC}$ is the ionic calcium current.
- $J_{pump}$ is the current caused by the membrane-bound pumps (to keep Ca++ concentration low on the inside).
- $J_{rel}$ and $J_{up}$ keep the ion concentration regulated in the EER.
  - $J_{up}$ Uptake to the endoplasmic reticulum.
  - $J_{rel}$ release from the ER stores.
- $J_{buff}$ is calcium buffering (molecules that can bind Ca++). This yields a neutral charge, such that the Ca++ effectively leaves the cell.
- $J_{drift}$ is the diffusive current $D\nabla^2[Ca^{++}]$.
- $J_{leak}$ is the typical leak out of the membrane.
