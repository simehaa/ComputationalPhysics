# Part 2, Hodgkin-Huxley Model

## 2.1 HH Model of Action Potential, AP (Sterratt 3.1)
So far we've only considered passive membranes
* Ohmic (resistive) currents $I_m$, $R_m$
* Capacitive currents $I_c$, $R_c$
Responds passively until action potential $V_{m, \theta} is reached. HH Model discusses the model of the action potentials.

We need some key features:
* Rapid polarization (increase V)
* Followed by rapid repolarization
* Often hyperpolarization
* Propagation down the axons with little loss of shape

#### HH-Model
One of the first *qualitatively* (predicted values) successful models in neuroscience. Predicted **shapes** and **voltages** (based on experimental recordings), and also predicted **propogation speed** (fairly accurately).

The HH-model is a circuit model (just like RC but a little more sophisticated), and the formalism is still in use.

## 2.2 Equivalent Circuit of Single Compartment HH Model
Have a membrane current which we can write as:

$$
I = I_c + I_L + I_{Na} + I_K
$$

The two first terms $I_c$ and $I_L$ are the passive components, defined as the capacitive current

$$
I_c = c_m \frac{dV}{dt}
$$
and the "leaking" current
$$
I_L = \bar{g}_L(V-E_m)
$$
- Same as the RC model.
- These two predict the passive response.
  - Passive response is everything below the AP-threshold.
- $\bar{g}_L$ indicates that it is constant.

The active channels are described in a similar way to the passive ones, though the conductances $g$ are not constant:
$$
I_{Na} = g_{Na}(V, t)(V-E_{Na})
$$
$$
I_K = g_K(V, t)(V-E_K)
$$
- The mechanisms behind AP's
- Time and Voltage dependant.

#### Time and voltage dependant conductances $g$

Can rewrite $I_x = g_x(V,t)(V-E_x)$ as:

$$
g_x(V,t) = \bar{g}_x \cdot p_x(V,t)
$$

Constant times something that changes in $V$ and $t$. This $\bar{g}_x$ is equal to the conductance of the fully open channel. $p_x$ is the probability (between 0 and 1) that a single channel is fully open. So given that there are many channels, we model the $p_x$ as being the fraction of channels that are fully open.

## 2.3 Potassium Current (Sterratt 3.2.1)
HH-model has $I_K = g_K(V,t)(V-E_K)$. How does this fit to experiments?

In Experiments they:
- use a **Space Clamp** to short circuit the neuron (meaning that $V_m$ is the same everywhere).
- Let neuron rest
  - Observed that almost no K+ was recorded.

Try again:
- **Clamp** at some potential $V_h$ (holding potential), f.ex. $25mV$ above rest.
- Record how the K+ current develops in time
- Estimate conductance through knowing that $g(V,t)=I_K(V,t)/(V_h-E_K)$.

#### What we observe when we do this:
- The conductance $g$ *increases* with time,
- The conductance settles at some value that we call $g_{K,\infty} (V_h)$
  - The steady state conductance depends on the holding potential $V_h$
- The time it takes to reach the steady state also depends on $V_h$.

### What are the mechanics of this system ?
The idea of HH was that each channel is controlled by one or more **Gating Particles**.

They hypothosized that each of these gating particles were in an *open* or *closed* state.

##### For a channel to be open, all the gating particles must be in the open state.

They introduced a formalism for this:

A gating particle is described with a variable $n$.

This variable describes the probability that a gating particle is in an open state.

If a channel has $z$ gating particles, then the probability for the channel to be open is $n^z$.

HH found that $z=4$ gives the best agreement with the data.
$$
\Rightarrow g_K(V,t) = \bar{g}_K\cdot n(V,t)^4
$$

HH invisioned that the $n$ function can be decribed by the dynamics of two states $C$ and $O$. Two rates describe there two states $\alpha_n$ (the rate of closed to open), and $\beta_n$ (the rate of open to closed). This yields the first order kinetic equation:
$$
\frac{dn}{dt} = \alpha_n(1-n) - \beta_nn
$$
Alternatively, the equation can be written as:
$$
\frac{dn}{dt}=\alpha_n - (\alpha_n + \beta_n)n
$$
$$
=(\frac{\alpha_n}{\alpha_n + \beta_n}-n)(\alpha_n + \beta_n)
$$
$$
=\frac{n_{\infty}-n}{\tau_n}
$$
With $n_{\infty} = \alpha_n/(\alpha_n + \beta_n)$,  and $\tau_n = 1/(\alpha_n + \beta_n)$. These quantities are easier to measure.

Interpretation: $n$ approaches $n_\infty$ with a time constant $\tau_n$. These all depend on the systems potential $V_h$.

Def:
* $n_\infty$ is the open probability (at a given voltage) when you wait long enough ($t\rightarrow \infty$).

Note:
$n_\infty$ and $\tau_n$ are functions of $V_h$ and the parameters $\alpha_n$ and $\beta_n$ are functions of $V_h$.

General solution of the equation
$$
\frac{dn}{dt}=\frac{n_{\infty}-n}{\tau_n}
$$
For a given potential $V_h$ is:
$$
n(t) = n_\infty (V_h) - (n_\infty (V_h) - n_0)exp(-t/\tau_n(V_h)),
$$
where $n_0$ is what $n$ was before we clamped cell at $V_h$ (i.e. at $t=0$). These are functions of voltage $V$, but the voltage must be clamped at $V_h$ to record the values of $t\rightarrow \infty$.

### How to determine $n_\infty$ and $\tau_n$ ?
* Create an experiment:
  - Let the neuron rest at $V=V_{rest}$
    - Here we record zero potassium current.
    - We know that the potassium channels are closed
      - $n_0=0$
  - Switch potential to holding potential $V_h$
  - Record for various different holding potentials.
  - Do this to get the best fit to the $n_\infty$ and $\tau_n$ curves.

#### Details on this;
We record the current $I_K$, given by:
$$
I_K = \bar{g}_K\cdot n^4 (V_h-E_K)
$$
From this, we implicitly know the conductance $g_K = \bar{g}_Kn^4 = I_K/(V_h-E_K)$

We record $I_K$ and know $(V_h-E_K)$.

With 4 gates set:

$$
n_\infty (V_h) = \left(\frac{g_{K, \infty}(V_h)}{\bar{g}_K} \right)^{1/4}
$$

Set $\tau_n$ to fit:
$$
I_K = \bar{g}_K n^4_{\infty} \left( 1-e^{-t/\tau_n} \right)^4
$$
to the data


### Finally,
If we have $n_\infty$ and $\tau_n$, we can compute:
$$
\alpha_n = n_\infty/\tau_n
$$
and
$$
\beta_n = \frac{1-n_\infty}{\tau_n}
$$
and then we get:
$$
I_K = \bar{g}_K n^4(V-E_K)
$$
And
$$
\frac{dn}{dt} = \alpha_n (1-n) - \beta_n n
$$
The fits of the data yielded:
$$
\alpha_n = 0.01\cdot\frac{V+55}{1-e^{-(V-55)/10}}
$$
$$
\beta_n = 0.125 \cdot e^{-(V-63)/80}
$$


## 2.4 Sodium Current (Sterratt 3.2.2)
Sodium has three gates of one type and an additional gate which behaves differently. Three tend to open with increasing voltage and the other tends to close with increasing voltage.

- Unlike K+, the Sodium Na+ conductance/current is transcient
  - peaks and then decays to zero.
- When V is switches on to certain value $V$:
  - Something opens it: Activates the Na+ channel
  - Then something seems to close: Inactivates the Na+ channel

### HH introduced:
- Activation variable $m$ (as $n$ was for the potassium current)
- Inactivation variable $h$

The model is then:
$$
g_{Na} = \bar{g}_{Na} m^3h,
$$
where $m$ and $h$ are probabilites $m, h \in [0,1]$.

#### Summary:
Sodium Channel:
$$
I_{Na} - \bar{g}_{Na} m^3h(V-E_{Na})
$$
has:
$$
\frac{dm}{dt} = \alpha_m (1-m) - \beta_mm
$$
and
$$
\frac{dh}{dt} = \alpha_h (1-h) - \beta_h h
$$
have the same form, but with different factors:
- $\alpha_m = 0.1 \cdot \frac{V+40}{1-exp(-(V+40)/10)}$
- $\beta_m = 4 \cdot exp(-(V+65)/18)$
- $\alpha_h = 0.07 \cdot exp(-(V+65)/20)$
- $\beta_h = \frac{1}{1+exp(-(V+35)/10)}$

This is assuming theres enough gates to say that the probability of gates being open is equal to the fraction of open gates.


## 2.5 Leak Current (Sterratt 3.2.3)
#### Leak conductance
* Same as in RC-model
* $I_L = \bar{g}_L (V-E_L)$
* $I_L$ will drive the memb potential towards $V=E_L$
* When $I_L$ is the *only* current, $E_L$ is identical to the resting potential
* When there are other ion channels, they can in principle be open during rest and $E_L$ will not be identical to the resting potential
* $I_L$: due to a mixture of ions. Kind of summarizes everything we don't model in further detail (e.g. $Na^+/K^+$ pumps $\neq$ a lot of unspecified ion channels)

## 2.6 Full HH-Model
$$C_m \frac{dV}{dt} = -\bar{g}(V-E_L) -\bar{g}_K n^4(V-E_K) -\bar{g}_{Na} m^3h(V-E_{Na}) + I_{Stim}$$

Parameters:
* $C_m = 1 \mu F/cm^2$
* $E_{Na} = 50 mV$
* $E_K = -77 mV$
* $E_L = -54.4 mV$
* $\bar{g}_{Na} = 120mS/cm^2$
* $\bar{g}_{K} = 36 mS/cm^2$
* $\bar{g}_{L} = 0.3 mS/cm^2$
Kinetics of *m, h, n* as described earlier.

## 2.7 Simulations of AP's (Sterratt 3.3)
* HH simulating space-clamped AP's: good match with experiments
* $Na^+$-channels open first and $K^+$-channels follow
* $K^+$-currents last longer: gives *hyperpolarization*
* HH explains refractory period: after AP state variables must reset before the neurons are ready to fire again
* Absolute refractory period: impossible for neuron to fire new AP
* Relative: more difficult to fire AP, requires stronger input stimulus
* HH model predicts an AP-propagation speed of $18.8 m/s$ at $18.5^\circ C$, which is close to measured speed of $21.2 m/s$

## 2.8 Effects of Temperature (Sterratt 3.4)
* Rates of $\alpha$ and $\beta$ depend on temperature
* Quantified by $Q_{10}$-factor, where $Q_{10}$ is the rate at temp. $T+10^\circ C$ divided by the rate at $T$
* $\alpha(V,T_2) = \alpha(V,T_1)Q_{10}$
* $Q_{10}$ depends on T, however, in HH a constant $Q_{10} = 3$ is a good approximation
* Also conductances for fully open channe;s depend on T
* For $\bar{g}_X$ the $Q_{10}$-factor is about $1.2 - 1.5$

## 2.9 HH-type Models
* Many neurons exits with different firing properties
    * Not all well described by the HH model
    * Many types of ion channels exits
        * $Ca^{2+}$-channels
        * $K^+$-channels
        * $Na^+$-channels
* Good news: HH type formalism applies to many of these
* HH type formalism means:

$$I_X = \bar{g}_X m^a h^b(V-E_X),$$

where 
* $\bar{g}_X$ is the conductance for fully open channel
* $a$ is the number of activation gates
* $b$ is the number of inactivation gates
* $E_X$ is the reversal potential for ion channel $X$



```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LIF(noise=False):
    h = 0.1e-3
    T = 1.0
    N = int(T/h + 1)
    tm = 10e-3 # ms
    R = 0.04e9 # G Ohm
    theta = 15e-3 # mV
    t = np.linspace(0,T,N)
    V = np.zeros(N)
    Ni = 101
    I = (np.linspace(0,1000,Ni))*1e-12 # np.ones(N)*400e-12
    if noise:
        I += np.random.uniform(0,22,Ni)*1e-12

    f = np.zeros(Ni)

    for i in range(Ni):
        for n in range(N-1):
            V[n+1] = V[n] + h/tm*(-V[n] + R*I[i])
            if V[n+1] > theta:
                V[n+1] = 0
                f[i] += 1
    return I, f/T


I, f = LIF()
plt.plot(I, f, "b--", label="I")
I, f = LIF(noise=True)
plt.plot(I, f, "k.", label="I with Noise")
plt.legend()
plt.grid()
plt.show()
```
