# Part 1, Electrical Properties of Neuron
Modelling neurons - since early 1950's.

## 1.1 Neural Membrane
* Neurons are confined by bilipid layer ($\approx 5 nm$ thick)
* Resting membrane potential ($\approx -(60 - 70) mV$), more negative on the inside
* Key ions
    * $K^+$ (more on the inside)
    * $Na^+$ (more on the outside)
    * $Cl^-$
    * $Ca^{2+}$
* Ion channels: pores in the libid bilayer
    * Protein tunnels
    * Different channels for different ions
* Types of channels
    1. Active channels: exist in *open* or *closed* states. Openness depend on
        * Membrane potential
        * Ion concentrations
        * Brand molecules
    2. Passive channels: permeability does not change with state
    3. Ion pumps: requires energy

## 1.2 Ion Movement in Neurons (Sterratt 2.2)
Two fundamental processes: *electrical drift* and *diffusion*.
### 1.2.1 Electrical Drift
$$ F_E = qE, $$
$$ q = Ze, $$
$$ F_E = md + bv \approx bv_{exit}, $$
where $b$ is a friction constant
* Mobility: $\mu = \frac{v_{drift}}{F}$
* In electrical field: $v_{drift} = \mu F_E = \mu q E = -\mu q \frac{dV}{dz}$
* Particle flux: $j_{x, drift} = C_x v_{x, drift} = -\mu_X C_X Z_X \frac{dV}{dz}$
* Einstein relation: $\mu_X = \frac{D_X}{k_B T}, with $D_X$ as a diffusion coefficient

$$ j_{X, drift} = - \frac{D_X e}{k_BT} C_x Z_X \frac{dV}{dz} \Bigg[\frac{\#}{m^2s}\Bigg] $$
In chemical units:
$$ J_{X, drift} = - \frac{D_X F}{RT}[X] Z_X \frac{dV}{dz} \Bigg[\frac{mol}{m^2s}\Bigg] $$

### 1.2.2 Diffusion
* All concentration gradients give a diffusive flux:
$$ J_{X, diff} = - D_X \frac{d[X]}{dy} $$

### 1.2.3 Electrodiffusion
* Total flux:
$$ J_X  = J_{X, diff} + J_{X, drift}$$
which is also the *Nernst-Planck* equation:
<center>
    <div class="alert alert-block alert-info">
    \begin{align}
    J_X = -D_X \left(\frac{d[X]}{dy} +  \frac{Z_X F}{RT}[X]\frac{dV}{dz}\right)
    \end{align}
    </div>
</center>

* Often, one is interested in electrical current:
$$ I = I_{Na} + I_K + I_{Cl} = \sum_{X\in Ion} F Z_X J_X $$
$$ I = FJ_{Na} + FJ_K - FJ_{Cl} $$

#### IV-relations
* Ohmic current: $I = const. V$ (flow of current outside **or** inside)
* Quasi-Ohmic current: $I = const. (V-V_{rest})$ (passive channel currents across membrane)
* Non-Ohmic current $I \neq const. (V- V_{rest})$ (active ion channels)


## 1.3 Nernst Potential
* Semipermeable membrane allowing only $K^+$ through,
    * final equilibrium state with more $K^+$ on the RHS
    * non-zero equilibrium $\Rightarrow$ membrane potential
* In this equilibrium: el. drift flux $=$ diffusion current
* What is equilibrium potential?
Set $J_X$ in *Nernst-Planck* equation to 0:

<center>
    <div >
    \begin{align}
     -D_X \left(\frac{d[X]}{dy} +  \frac{Z_X F}{RT}[X]\frac{dV}{dz}\right) &= 0 \\
      \frac{1}{[X]} \frac{d[X]}{dy} &= -\frac{Z_X F}{RT}[X]\frac{dV}{dz} \\
      -dV &= \frac{RT}{Z_X F [X]} d[X] \\
      -\int_{V_{in}}^{V_{out}} dV &= \frac{RT}{Z_X F [X]} \int_{[X]_{in}}^{[X]_{out}} \frac{1}{[X]} d[X] \\
      V_{in} - V_{out} &= \frac{RT}{Z_X F} \ln{\frac{[X]_{out}}{[X]_{in}}}
    \end{align}
    </div>
    <div class="alert alert-block alert-info">
    \begin{align}
        E_X = \frac{RT}{Z_X F} \ln{\frac{[X]_{out}}{[X]_{in}}}
    \end{align}
    </div>
</center>

$E_X$ is also known as the *Nernst potential*.

## 1.4 Goldman-Hodgkin-Katz Equations (GHK)
Nernst equation for single ion species when drifts in one direction $=$ diffusion in the other direction. For many ion species: the equilibrium consists of fluxes of several ions in several directions. A prediction of *resting membrane potential* requires quantitative modelling of the various currents.
$$ I_X = P_X Z_X F \frac{Z_X F V}{RT}\left( \frac{[X]_{in} - [X]_{out} e^{-Z_X FV/RT}}{1 - e^{-Z_XFV/RT}} \right) $$

* Membrane permeability of ion X: $P_X [cm/s]$ 
    * $J_X = -P_X([X]_{in} - [X]_{out})$
    * $P_X \sim D_X$
    * Derived by requiring
$$ \sum_{X \in ion} i_X = 0 $$
$$ \sum_{X \in ion} FZ_Xj_X = 0 $$
$$ \sum_{X \in ion} FZ_X\left( -P_X\frac{d[X]}{dz} - P_X \frac{Z_XF}{RT}[X] \frac{dV}{dz}  \right) = 0 $$
#### IV-characteristics
Found by plotting I vs. V with the GHK equation.
* $K^+$: I is mostly $>0$, which favours an outward current
* $Ca^+$: I is mostly $<0$, which favours an inward current

#### Equilibrium
Found by setting $I = 0$:
$$ I_{Ca} + I_{Na} + I_K + I_{Cl} = 0 $$
and neglect $I_{Ca}$. The GHK equation yields:
$$ E_M = \frac{RT}{F} \ln{\left( \frac{P_K[K^+]_{out} + P_Na[Na^+]_{out} + P_Cl[Cl^-]_{in}}{P_K[K^+]_{in} + P_Na[Na^+]_{in} + P_Cl[Cl^-]_{out}}\right)}$$

Note:
* $Cl^-$ swaps in-out concentrations due to negative charge
* Only relative permeabilities matter for squid-giant axon:
    * $P_K : P_{Na} : P_{Cl} = 1 : 0.03 : 0.1$
    
$$\Rightarrow E_m \approx E_K$$

since $P_K \gg P_{Na}, P_{Cl}$

#### Quasi-Ohmic Approximation
In neuron modelling it is common to assume $I_X = g_x (V-E_X)$, where $(V-E_X)$ is the driving force and $g_x = \frac{1}{R_X}.

## 1.5 RC-circuit Neuron (Sterratt 2.5-2.6)
* $I_e$: electrode current
* $I_C$: capacitive current (the missing piece)
$$ I_C = \frac{dq}{dt} = C_m \frac{dV}{dt} $$

* $I_m$: ionic current(s)

Simplest case: consider piece of membrane with quasi-ohmic channels, ions x, y, z:

<center>
    \begin{align}
    I_m &= I_x + I_y + I_z \\
        &= g_x (V-E_x) + g_y (V-E_y) + g_z (V-E_z)\\
        &= (g_x + g_y + g_z)\left( V - \frac{g_xE_x + g_yE_y + g_zE_z}{g_x + g_y + g_z} \right) \\
        &= g_m (V - E_m) \\
        &= \frac{1}{R_m} (V-E_m)
    \end{align}
</center>

#### Kirchoff's Current Law 
Consider the whole cell with area $a$:

<center>
    \begin{align}
    aI_C + aI_m - I_e &= 0 \\
    I_C &= \frac{I_e}{a} - I_m \\
    C_m \frac{dV}{dt} &= \frac{E_m - V}{R_m} + \frac{I_e}{a}
    \end{align}
</center>
we obtain a DE which can be solved. For a current injection during $t \in [0,t_e]$, the analytical solution is

* Before and during current injection:
<center>
    \begin{align}
    V(t) &= E_m + \frac{R_mI_e}{a} (1 - e^{-t/R_mC_m})\\
    V(0) &= E_m\\
    V(t\to \infty) &= E_m + \frac{R_mI_e}{a}
    \end{align}
</center>

* After current injection:
<center>
    \begin{align}
    V(t) &= E_m + (V(t_e) - E_m) e^{-(t-te)/R_mC_m}\\
    V(t_e) &= V(t_e)\\
    V(t\to \infty) &= E_m
    \end{align}
</center>

Important parameters:
* Membrane time constant: $\tau_m = R_mC-m$ (typically $5-50ms$)
* $R_{in} = \frac{V_\infty - E_m}{I_e}$
