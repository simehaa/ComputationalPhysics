## 5.1 Ion Channel Structure and Function (5.1-5.2 in Sterratt)
Have ATCG which makes ca. 20 different proteins. Some are fatty (not polar) and are hydrophobic. This is how the ion channels are designed.

- Ion channels are made of proteins going through the membrane.
- Ion channels are constructed by **protein subunits**
  - Split the ion channel into sub-units
  - For a particular subunit called S4, it responds to voltage (Voltage Sensing Domain).
    - Turns out the excitability of these (too much change in typical excitability of S4) may cause hallucinations and schizophrenia.
    - If you're not able to dampen the excitability (pre-pulse inhibition, etc at Ullevaal Sykehus).
    - See work by Ole Andreassen at UiO.  
  - Gating charge in S4 is affected by membrane voltage.

The human genome was mapped out in 2003, and about ~140 ion channel types have been identified.
We have different schemes for naming these ion channels:
- IUPHAR: Chemical symbol for the principal permeating ion (K, Na, Cl etc.), followed by a subscript indicating the activator (Ca, V, etc. whether it's open or not).
- Gene nomenclature: Set by comittee :/
- Ad hoc nomenclature: Sometimes used when gene combination or ion combination is not known.

This is how we can simplify the huge zoo of ion channels:
- If you want something inhibitory potassium or cloride (RP which is negative ~-80, draws the resting potential towards hyperpolarization).
- And if you want something excitory use sodium or calcium reversal potential which is positive (draws the resting potential towards +, depolarization).

It's not that bad.

We also have websites which catalogue the different ion channels:

  - icg.neurotheory.ox.ac.uk

Makes it easier to compare work.
Remember that the membrane is hydrophilic on both ends. The middle of the lipid layer is hydrophobic.

## 5.2 Experimental Techniques
Coolest:
#### Single channel recordings with a patch-clamp electrode.
Developed by sachman and mayer in the 80's. Nobel prize in 91.
**How to**
Use very sharp hollow electrode, put it next to the membrane, and suck the membrane into the tip.
Get a very tight seal such that the only current measured is the current which goes through the ion channel. Typically know the density of channels. It is possible to get only one channel but it is unlikely. Usually get multiple channels from which you can assert the affect of single ones.
Typically do this in slices, can see the neuron under the microscope... People are very good at doing this.

**Cont.**
Opening and closing of channels are stochastic.
The chances of being open and closed are not stochastic however. Channels are either open or closed, and the probability of being open varies with voltage.

#### Channel Blockers:
- Pharmacological compounds which block particular channel currents:
  - TTX: Block NA+ channels
  - TEA: Block some K+ channels

##### Channel isolation by mRNA insertion
Instead of blocking all the other channels, you just insert the channel you are interested in.

### Gative Current:
Know that S4 responds to voltage.
Moves physically, has a charge $\Rightarrow$ produces a little current (gative current, analagous to capacative gurrent)
Current from the gate opening, can be measured.

## 5.3 Modeling ensambles of voltage-gated Ion Channels (5.5 in Sterratt)
Model these as continuous models: Probabilities of being open.
It is the average affect of channels that is typically measured.

#### The most used model is the gating-particle model:
- Ion current modeled in HH-style by means of *activating* (a) and *inactivating* (b) particles:
$$
I(t) = \bar{g} a^x b^y (V-E_r)
$$

This is what we saw for the sodium and potassium channels.
The first order differential equations (ODE's) describing dynamics of a and b are then (just as we did for HH):
$$
I_A = \bar{g}_A a^3b(V-E_A)
$$

This is for the additional (to Na and K), A current (potassium).

This was more suited for explaining the AP's for the crab axons (rather than squid).
HH was what was called type two: No firing but then firing at a particular current.
The integrated fire model was type 1. Means a gradual firing.

#### Addition of $I_A$ changed HH from type 2 to type 1

type 1: you increase the current and get arbitrarily long times between spikes (e.g. frequencies of 0.001).
type 2: you suddenly get 80Hz at a certain current. Zero before that.

#### In thermodynamic models:
The form of the rate coefficients are derived from the **transition state theory**, rather than simply fitted:
$$
x_\infty = \frac{1}{1+e^{-(V-V_{1/2})/\sigma}}
$$
$$
\tau_x \rightarrow \tau_x(V_i V_{1/2}, \delta)
$$

(eqn's 5.2 and 5.3 in Sterratt). This is not very important in practice. Don't think about this.
