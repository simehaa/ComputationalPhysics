# Part 3, Multicompartmental Models and Cable Equation (Sterratt 2.8)
* Single compartment models
    * What we've studied so far
    * Assume that neuron is isopotential: V is the same everywhere
    * This is not true, but only a simplified model
* Multi compartmental models
    * Split neuron into cylinders
    * Connected with resistors
* Look at stick (part of dendrite or axon)
    * Split into cylinders
        * Length *l*
        * Diameter *d*
        * Surface area $a = \pi ld$
        * Cross section $\pi d^2/4$
* Current can flow:
    1. Accross membrane
    2. Axially inside neuron
    3. Axially outside neuron
        * Extracellular resistance is much smaller than intracellular resistance
        * Common assumption: extracellular medium is grounded $V=0$ everywhere outside (It is generally not zero, but much smaller than on the inside)
Kirchoff: 

$$
I^m_j + I_{j, j+1} - I_{j-1, j}= 0
$$
 
Membrane current $I_j^m$ as before, 

$$
I_j^m = I_j^{cap} + I_j^{ion} + I_j^{stim} = -\pi dl C_m \frac{dV}{dt} - \pi dl \frac{V_j - E_m}{R_m} + I_j^{stim}
$$


* Axial currents (New)
    * Axial resistivity: $R_a [\Omega cm]$ 
    * Axial resistanceL $\frac{R_a l}{\pi d^2/4} [\Omega]$
* Ohm's law:
$$
I_{j-1,j} = \frac{V_{j-1} - V_j}{4R_a / (\pi d^2)}
$$
* Insert membrane currents and axial currents in Kirchoff's equation:

<center>
    <div class="alert alert-block alert-info">
    \begin{eqnarray}
        C_m \frac{dV}{dt} = \frac{E_m - V_j}{R_m} + \frac{d}{4R_a} \Bigg( \frac{V_{j+1} - V_j}{l^2} - \frac{V_j -V_{j-1}}{l^2} \Bigg) + \frac{I_{stim}}{\pi d l}
    \end{eqnarray}
    </div>
</center>
* Fundamental equation for multicompartment modelling. We can add additional membrane currents to the right hand side (e.g. HH).
* Multicompartment modelling can be solved with dedicated software such as *neuron* (popular option) or *genesis*

## 3.2 Cable Equation (Sterratt 2.9)
* Multicompartment models must typically be solved numerically
* Cable equation allows for analytical understandment of electrical properties of dendrites
* Idea: infinitely many small compartments with length $\delta$
* Current injected per unit length: $I_e (x,t) = \frac{I_{ej}}{\delta x}$

Rewrite multicompartmental model ($l \rightarrow \delta x$):
$$
C_m \frac{dV(x,t)}{dt} = \frac{E_m - V(x,t)}{R_m} + \frac{d}{4R_a} \frac{1}{\delta x}\Bigg( \frac{V(x+\delta x,t) - V(x,t)}{\delta x^2} - \frac{V(x,t) -V(x-\delta x,t)}{\delta x^2} \Bigg) + \frac{I_{e}(x,t)}{\pi d}
$$
$$
\Rightarrow \frac{\delta^2 V}{\delta x^2}
$$

Introduce two new variables
1. Membrane time constant: $\tau_m = R_m C_m$
2. Electro length constant: $\lambda = \sqrt{\frac{d R_m}{4 R_a}}$

We obtain the cable equation:

<center>
    <div class="alert alert-block alert-info">
    \begin{eqnarray}
    \tau_m \frac{\partial V}{\partial t} = E_m - V + \lambda^2 \frac{\partial ^2 V}{\partial  x^2} + \frac{I_{e}R_m}{\pi d}
    \end{eqnarray}
    </div>
</center>

#### Boundary Conditions 

* Sealed-end BCs (no axial current)
$$ \frac{\partial V}{\partial x}=0 $$
* Killed-end BCs (no membrane potential)
$$ V = 0 $$


## 3.3 Steady-State Solutions of Cable Equation (Sterratt 2.9)
* Current injected at sealed end of semi-infitie cable:
$$ 0 = E_m - V + \lambda^2\frac{d^2 V}{dx^2} + (\text{current at } x=0) $$
* Introduce 
$$ \hat{V} = V - E_m \longrightarrow \frac{\partial \hat{V}}{\partial x} = \frac{\partial V}{\partial x} $$
DE. now becomes
$$ \frac{\partial^2 \hat{V}}{\partial x^2} - \frac{1}{\lambda^2} \hat{V} = 0$$
Solution:
$$ \hat{V}(x) = Ae^{-x/\lambda} + Be^{x/\lambda}$$
with 
* $B = 0$ since $\lim_{x\to \infty}\hat{V}= 0 $
* $A = \hat{V}(0)$

Solution:
$$ \hat{V} (x) = V(x) - E_m$$
Thus:

<center>
    <div class="alert alert-block alert-info">
    \begin{align}
    V(x) = E_m + (V_0 - E_m)e^{-x/\lambda}
    \end{align}
    </div>
</center>

* Input resistance: $R_\infty = \frac{\Delta V}{I_e} = \frac{V_0 - E_m}{I_e}$
* Many introduce: 
    * $r_m = \frac{R_m}{\pi d} \Rightarrow \frac{R_m}{\text{circumference}}$, membrane resistance per inverse unit length
    * $r_a = \frac{4R_a}{\pi d^2} \Rightarrow \frac{R_a}{\text{cross-section area}}$, axial resistance per unit length
    
$$\lambda = \sqrt{\frac{dR_m}{4 R_a} } = \sqrt{\frac{r_m}{r_a}}$$

The smaller the membrane resistance is compared to axial resistance, the shorter $\lambda$ will be (analogous to a leaky garden hose)

* What is the input resistance?

<center>
    <div>
    \begin{align}
    I_e &= I_a (x=0) \\
        &= \frac{1}{r_a}\frac{V(\delta x) - V(0)}{\delta x} \\
        &= \frac{1}{r_a} \frac{\partial V}{\partial x} \Bigg|_{x=0} \\
        & = \frac{1}{r_a} \frac{d}{dx} \left( \hat{V}(0) e^{-x/\lambda} \right) \Bigg|_{x=0} \\
        & = -\frac{1}{r_a}\frac{1}{\lambda} \\
     R_\infty &= \frac{\hat{V}(0)}{|I_e|} \\
        &= \frac{\hat{V}(0)}{\frac{1}{r_a \lambda} \hat{V}(0)} \\
        &= r_a \lambda \\
        &= r_a \sqrt{\frac{r_m}{r_a}} \\
        &= \sqrt{\frac{4R_mR_a}{\pi^2 d^3}}
     \end{align}
    </div>
</center>


## 3.4 Temporal Solutions of the Cable Equation (Sterratt 2.9)
General solution:
$$V(x,t) = C_0 e^{-t/\tau_0} + C_1 e^{-t/\tau_1} + C_2 e^{-t/\tau_2} + ...$$
with $\tau_0 > \tau_1 > \tau_2 > ...$ and $\tau_0 = \tau_m = R_mC_m$ (membrane time constant)

* In figure (2.17b) the **membrane potential** the same at all posistions after about $7ms$, but not yet back at **resting membrane potential** $\tau_m = R_mC_m = 1\Omega m^2 0.01 F/m^2 = 10 ms$
